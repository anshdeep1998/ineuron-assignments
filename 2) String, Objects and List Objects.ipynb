```python
for i in range(5): # i traverses 0,1,2,3,4
    for j in range(i+1):#j traverses 1,2,3,4,5 (bcoz 0+1,1+1,2+1,3+1,4+1)
        print("* ",end="")
    print("\r")#adding this line, always starts new line after printing
    
#Bcoz in the 2nd "j" loop it prints "." and after printing ".", when 2nd loop ends, control goes to 1st loop
#and inside 1st loop, print("\r") takes to next line

for i in range(4):#0,1,2,3,
    for j in range(i,4):# range(0,5) prints 5 dots, then range(1,4) prints 3 dots in next row, then range(2,4) prints 2 dots
                        #in next row, then range(3,4) prints 1 dot in next row, then range(4,4) prints 0 dot finally
        print("* ",end="")
    print("\r")
   
```
